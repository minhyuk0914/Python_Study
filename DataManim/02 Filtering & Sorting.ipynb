```python
import pandas as pd
```

```python
# Q20 데이터를 로드 
DataUrl = 'https://raw.githubusercontent.com/Datamanim/pandas/main/chipo.csv'
df = pd.read_csv(DataUrl)
type(df)
```

```python
# Q21 quantity컬럼 값이 3인 데이터를 추출하여 첫 5행을 출력
df.loc[df["quantity"]==3].head(5)
```

```python
# Q22 quantity컬럼 값이 3인 데이터를 추출하여 index를 0부터 정렬하고 첫 5행을 출력

# reset_index(drop= , inplace=) 

df.loc[df["quantity"]==3].head().reset_index(drop=True)
```

```python
# Q23 quantity, item_price 두 개의 컬럼으로 구성된 새로운 데이터 프레임을 정의하라
df[["quantity", "item_price"]]
```

```python
# Q24 item_price 컬럼의 달러표시 문자를 제거하고
#     float 타입으로 저장하여 new_price 컬럼에 저장
df["new_price"] = df["item_price"].str[1:].astype('float')
df["new_price"].head()
```

```python
# Q25 new_price 컬럼이 5이하의 값을 가지는 데이터 프레임을 추출하고, 전체 갯수 구하기
len(df.loc[df["new_price"]<=5])
```

```python
# Q26 item_name명이 Chicken Salad Bowl 인 데이터 프레임을 추출하라고 index 값을 초기화하기
df.loc[df["item_name"]=="Chicken Salad Bowl"].reset_index(drop=True)
```

```python
# Q27 new_price값이 9이하이고, item_name 값이 Chicken Salad Bowl인 데이터 프레임 추출
df.loc[(df["new_price"]<=9) & (df["item_name"]=="Chicken Salad Bowl")]
```

```python
# Q28 df의 new_price 컬럼 값에 따라 오름차순으로 정리하고 index를 초기화하기
df.sort_values("new_price").reset_index(drop=True)
```

```python
# Q29 df의 item_name 컬럼 값 중 Chips 포함하는 경우의 데이터를 출력

# str.contains("&&&") - &&&를 포함한 문자열 찾기

df.loc[df.item_name.str.contains('Chips')]
```

```python
# Q30 df의 짝수번째 컬럼만을 포함하는 데이터프레임을 출력
df.iloc[:, ::2]
```

```python
# Q31 df의 new_price 컬럼 값에 따라 내림차순으로 정리하고 index를 초기화

# ascending=True - 오름차순 - default 값
# ascending=False - 내림차순 

df.sort_values("new_price", ascending=False)
```

```python
# Q32 df의 item_name 컬럼 값이 Steak Salad 또는 Bowl 인 데이터를 인덱싱하라
df.loc[(df.item_name == "Steak Salad") | (df.item_name == "Bowl")]
```

```python
# Q33  df의 item_name 컬럼 값이 Steak Salad 또는 Bowl 인 데이터를 데이터 프레임화 한 후,
#      item_name를 기준으로 중복행이 있으면 제거하되 첫번째 케이스만 남겨라

# drop_duplicates(&&&, keep=first(첫번째) or last(마지막), False(전체 제거)) - 중복행이 있으면 제거하고 첫 번째 값만 남기기 - default

Q33_df = df.loc[(df.item_name == "Steak Salad") | (df.item_name == "Bowl")]
Q33_df.drop_duplicates('item_name')
```

```python
# Q34 df의 item_name 컬럼 값이 Steak Salad 또는 Bowl 인 데이터를 데이터 프레임화 한 후,
#          item_name를 기준으로 중복행이 있으면 제거하되 마지막 케이스만 남겨라
Q34_df = df.loc[(df.item_name == "Steak Salad") | (df.item_name == "Bowl")]
Q34_df.drop_duplicates('item_name', keep='last')
```

```python
# Q35 df의 데이터 중 new_price 값이 new_price 값의 평균값 이상을 가지는 데이터들을 인덱싱하라
df.loc[df.new_price >= df.new_price.mean()]
```

```python
# Q36 df의 데이터 중 item_name의 값이 Izze 데이터를 Fizzy Lizzy로 수정
df.loc[df.item_name == "Izze", 'item_name'] = 'Fizzy Lizzy'
df
```

```python
# 이후 문제들에서 choice_description 에 Nan 값이 존재하여
# 이를 fillna("Nodata", inplace=True)를 통해 Nan값을 변경하고 저장하여 사용
df.choice_description.fillna("Nodata", inplace = True)
df
```

```python
# Q39 df의 데이터 중 choice_description 값에 Black이 들어가는 경우를 인덱싱하라df
df[df.choice_description.str.contains("Black")]
```

```python
# Q40 df의 데이터 중 choice_description 값에 Vegetables 들어가지 않는 경우의 갯수를 출력
len(df.loc[~df.choice_description.str.contains("Vegetables")])
```

```python
# Q41 df의 데이터 중 item_name 값이 N으로 시작하는 데이터를 모두 추출

# startswith() / endswith()

df[df.item_name.str.startswith("N")]
```

```python
# Q42 df의 데이터 중 item_name 값의 단어갯수가 15개 이상인 데이터를 인덱싱하라
df[df.item_name.str.len() >= 15]
```

```python
# Q43 df의 데이터 중 new_price값이 lst에 해당하는 경우의 데이터 프레임을 구하고,
#     그 갯수를 출력하라
lst =[1.69, 2.39, 3.39, 4.45, 9.25, 10.98, 11.75, 16.98]

display(df.loc[df.new_price.isin(lst)])

print(len(df.loc[df.new_price.isin(lst)]))
```
